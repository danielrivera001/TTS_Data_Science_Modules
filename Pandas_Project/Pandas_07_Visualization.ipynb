### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
url='https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo=pd.read_csv(url, sep='\t')
```

### Step 4. See the first 10 entries

```python
chipo.head(10)
```

### Step 5. Create a histogram of the top 5 items bought

```python
#Specify your x-axis
x=chipo.item_name
letter_counts= Counter(x)
#Create a data frame of the counters of x
df = pd.DataFrame.from_dict(letter_counts, orient='index')
#sort the values of df in ascending order
df=df[0].sort_values(ascending = True)[45:50]
#plot a bar graph and include a title and labels
df.plot(kind='bar')
# Set the title and labels
plt.xlabel('Items')
plt.ylabel('Number of Times Ordered')
plt.title('Top 5 Most ordered Chipotle\'s Items')
# output the plot
plt.show()
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
# create a list for all of the priced items
chipo.item_price = [float(value[1:-1]) for value in chipo.item_price] # strip the dollar sign and trailing space

# use the groupby function to group them by order id. Take the sum and call is orders
orders = chipo.groupby('order_id').sum()

#Establish a scatter plot using the plt.scatter function
plt.scatter(x = orders.item_price, y = orders.quantity, s = 50, c = 'green')

# Specify the axis label and title yoour plot
plt.xlabel('Order Price')
plt.ylabel('Items ordered')
plt.title('Number of items ordered per order price')
plt.ylim(0)
```

```python
#Question: Plot a histogram of the least ordered Chipotle items
#Specify your x-axis
x=chipo.item_name
letter_counts= Counter(x)
#Create a data frame of the counters of x
df = pd.DataFrame.from_dict(letter_counts, orient='index')
#sort the values of df in ascending order
df=df[0].sort_values(ascending = False)[45:50]
#plot a bar graph and include a title and labels
df.plot(kind='bar')
# Set the title and labels
plt.xlabel('Items')
plt.ylabel('Number of Times Ordered')
plt.title('Bottom 5 Least ordered Chipotle\'s Items')
# output the plot
plt.show()
```
